```python
#----------------------PASEO ALEATORIO 1 DIMENSIN--------------------------
import numpy as np
import matplotlib.pyplot as plt
def Randomwalk1D(n): #n es el nmero de simulaciones deseadas
   x = 0
   y = 0
   xposition = [0] # empezamos del punto cartesiano (0,0)
   yposition = [0] 

   for i in range (1,n+1):
       step = np.random.uniform(0,1)

       if step < 0.5: # is es menos de 0.5 movemos hacia ariba 
           x += 1
           y += 1  

       if step > 0.5: # si es mas de 0.5 movemos hacia abajo  
           x += 1
           y += -1 
 
       xposition.append(x)
       yposition.append(y)

   return [xposition,yposition]

Randwalk = Randomwalk1D(10000) # creamos un objeto para la variable Randomwalk1D y le damos un valor de 10000 a la n 
plt.plot(Randwalk[0],Randwalk[1],'r-', label = "Randwalk1D") # 'r-' pintamos el camino aleatorio de color rojo
plt.title("1-D Paseo aleatorio")
plt.show()


#---------------------PASEO ALEATORIO 2 DIMENSIONES-----------------------

import numpy as np 
import pylab 
import random 
n = 3453 #n es el nmero de paso, cuantos ms pasaso tengamos mas complejo ser el grfico 
x = np.zeros(n) # x e y son matrices que almacenan las coordenadas de la posicin
y = np.zeros(n) 
direction=["NORTE","SUR","ESTE","OESTE"] # Las posibles direcciones 
for i in range(1, n): 
    step = random.choice(direction) #seleccionando una direccin aleatoria 
    if step == "ESTE": 
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
    elif step == "OESTE": 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1] 
    elif step == "NORTE": 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
pylab.title("2-D Paseo aleatorio")
pylab.plot(x, y) #GRFICO del paseo aleatorio.
pylab.show()

#----------------------PASEO ALEATORIO 3 DIMENSIONES-------------------------
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #libreria para graficas en 3D
N =545
R = (np.random.rand(N)*6).astype("int") #Pasos random de inicio
x = np.zeros(N) 
y = np.zeros(N)
z = np.zeros(N)
x[ R==0 ] = -1; x[ R==1 ] = 1 #asignacin de matrices a cada variable
y[ R==2 ] = -1; y[ R==3 ] = 1
z[ R==4 ] = -1; z[ R==5 ] = 1
x = np.cumsum(x) #La funcin cumsum() se usa para obtener una suma acumulada sobre un eje de datos o serie, es decir, suma los pasos para cada eje del plano.
y = np.cumsum(y)
z = np.cumsum(z)
plt.figure()
ax = plt.subplot(1,1,1, projection='3d')
ax.plot(x, y, z,alpha=0.9) #alpha establece los pasos invisibles.
ax.scatter(x[-1],y[-1],z[-1])
plt.show()
```
