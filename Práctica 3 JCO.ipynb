```python
#primero importamos la libreria numpy
import numpy as np
from scipy.stats import binom
#definimos las variables de la opción que vamos a usar y sus características
S = 100
E = 100
T = 2
n = 4
V = 0.2
L = 0.05
r = np.log(1+L)
#una vez definidas los datos, tenemos que calcular el tiempo, el factor de subida, bajada y la proabilidad
dT = T/n
u = np.exp(V*np.sqrt(dT))
d = 1/u
p = (np.exp(r*dT)-d)/(u-d)

Sn = [0]*(n+1)
print('u=', u, 'p=', p, 'dT', dT)
#Ahora generamos los precios esperados del activo subyacente a fecha de vencimiento.
#Sn será la lista de precios del activo subyacente a fecha de vencimiento.
for i in range(n+1):
  Sn[i]=S*u**(n-i)*d**i
#por último calculamos el payoff = 
  Payoff = [0]*(n+1)

for i in range(n+1): 
  if Sn[i] > E:
    Payoff[i] = Sn[i] - E
  else:
    Payoff[i] = 0

    prima = 0
#y ya por último calculamos la prima
for i in range(n+1):
  prima += Payoff[i]*binom.pmf(n-i, n, p)

prima = prima*np.exp(-r*T)

print('Prima estimada=', prima)

#sí precio de la acción en n es mayor que el precio de ejercicio, el valor de la opción será la diferencia entre ambos precios.
#sí el precio de la acción en n es menor que el precio de ejercicio, el valor de la opción será 0.
```
